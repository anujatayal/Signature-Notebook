```python
import cv2
import os
import numpy as np
import pandas as pd
import tensorflow as tf
#tf.enable_eager_execution()
import numpy.linalg as lin
import matplotlib.pyplot as plt
%matplotlib inline
from google.cloud.vision import types
from google.cloud import vision
import io,re
from difflib import SequenceMatcher
from tensorflow.keras.models import Model,load_model
import random
```

```python
def triplet_loss(y_true, y_pred, alpha = 0.2):  
    anchor, positive, negative = y_pred[:,0], y_pred[:,1], y_pred[:,2]
    print(anchor, positive, negative)
    pos_dist = tf.reduce_sum(tf.square(anchor - positive))
    neg_dist = tf.reduce_sum(tf.square(anchor - negative))
    basic_loss = pos_dist - neg_dist + alpha
    loss = tf.maximum(basic_loss, 0)
    return loss
def visionocr(imagePath, key):    
    os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = key #setting the environment variable
    #credentials = service_account.Credentials.from_service_account_file('apikey.json')
    client = vision.ImageAnnotatorClient()              #creating an Annotator client 
    with io.open(imagePath, 'rb') as image_file:
        content = image_file.read()                     #reading text from the image 
    image = types.Image(content=content)
    kwargs = {"image_context": {"language_hints": ["en"]}}
    response = client.document_text_detection(image, **kwargs)
    #response = client.text_detection(image)
    texts = response.text_annotations
    return(texts)

#model=load_model('weights_siamese.hdf5')
model =load_model('weights-siamese-v2.hdf5', custom_objects={'triplet_loss': triplet_loss})
def read_image(i):
    img1 = cv2.imread(data.loc[i,"img"])
    img1 = cv2.resize(img1, (220,150))
    img1 = np.moveaxis(np.array([cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)]),0,-1)
    return (img1/255.).astype('float32')
    return img1
    
def show_image(img1,i):
    plt.imshow(np.squeeze(img1))
    plt.title(i)
def read_image_path(path):
    img1 = cv2.imread(path)
    img1 = np.moveaxis(np.array([cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)]),0,-1)
    return (img1/255.).astype('float32')
    return img1
def img_to_encoding(img, model):
    x_train = np.array([img])
    embedding = model.predict_on_batch(x_train)
    return embedding
```

```python
#img_path="/home/anuja/Downloads/3.jpg"
img_path="/home/anuja/Downloads/8.png"
key = "/home/anuja/Dropbox/given_files/apikey.json"
def signature_extraction(img_path,key):
    df = visionocr(img_path, key)
    for text in df[1:]:
        if SequenceMatcher(None, text.description.upper().strip(), 'THANKS').ratio() > 0.8:
            #print("HI")
            thanks_vertices = [(text.bounding_poly.vertices[0].x,text.bounding_poly.vertices[0].y),
                                (text.bounding_poly.vertices[2].x,text.bounding_poly.vertices[2].y)]
        if SequenceMatcher(None, text.description.upper().strip(), 'RICHARD').ratio() > 0.8:
            #print("Hello")
            name_vertices = [(text.bounding_poly.vertices[0].x,text.bounding_poly.vertices[0].y),
                        (text.bounding_poly.vertices[2].x,text.bounding_poly.vertices[2].y)]
    img1=cv2.imread(img_path)
    img=img1[thanks_vertices[1][1]:name_vertices[0][1],name_vertices[0][0]:name_vertices[0][0]+300]
    return(img,name_vertices)

img=cv2.imread(img_path)
sig_image,name_vertices=signature_extraction(img_path,key)

def getTightBBox(cropImg):
    '''
        Removes the extra background while retaining signature
        Accepts grayscale image
        Returns bounding box of signature coordinates
    '''
    cropImg = cv2.GaussianBlur(cropImg,(3,3),0)
    cropImg = cv2.fastNlMeansDenoising(cropImg,None,10,7,27)
    cropImg=cv2.cvtColor(cropImg,cv2.COLOR_RGB2GRAY)
    
    cropImg = cv2.GaussianBlur(cropImg,(5,5),0)
    cropImg = cv2.adaptiveThreshold(cropImg,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,15,9)

    kernel = np.ones((3,3),np.uint8)
    inv = cv2.bitwise_not(cropImg)
    erosion = cv2.morphologyEx(inv, cv2.MORPH_OPEN, kernel)
    cropImg = cv2.bitwise_not(erosion)

    try:
        x_,y_ = np.where(cropImg==0)
        left_,right_ = min(y_),max(y_)
        top_,bottom_ = min(x_),max(x_)
    except:
        return
    return ((bottom_,top_,),(left_,right_))
```
